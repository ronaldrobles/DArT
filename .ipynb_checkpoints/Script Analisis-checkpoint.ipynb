### Obtener orden de muestras en el arbol

Se necesita este orden para plotear algunos resultados posteriores

```r


#cambiar extension .phy a .tre
# install.packages("ape")

library(ape)
tree <- read.tree("DArT2908.tre")

# ?ladderize :This function reorganizes the internal structure of the tree to get the ladderized effect when plotted
tree <- ladderize(tree, right = FALSE)

#First step is to filter out internal nodes from the the second column of the edge matrix:
is_tip <- tree$edge[,2] <= length(tree$tip.label)
is_tip[1:9]
#> [1]  TRUE FALSE FALSE  TRUE  TRUE FALSE  TRUE  TRUE
```

```r
ordered_tips <- tree$edge[is_tip, 2]

#Then you can use this vector to extract the tips in the right order:
listafinal <- tree$tip.label[ordered_tips]
listafinal[1:12]
#Ohhh.... ahora si sale en orden !!

write.table(listafinal, "orden arbol.txt", quote= F, sep="\t", row.names = F, col.names =F)
```

Como se eliminó un grupo grande de muestras, tambien habrán disminuido los marcadores polimorficos.
Contar las veces que aparece cada marcador (conteo de "1" en cada columna)

```r
#Esta es una prueba de la funcion que se usará

aj <- matrixTsilv[,4]
length(which(aj == "1"))
```

```r

frequency <- function(data){
	total <- c()	
	for (i in 1:length(data[1,])) { 
		count <- length(which(data[, i]== "1"))
		total[i] <- count}
	total
}
```

```r
# se prueba la funcion en un grupo pequeño
tot <- frequency(matrixTsilv[,1:20])
tot
tot  >= 1
# el resultado nos indica que columnas tienen uno o mas apariciones del marcador
```

```r
tot/2160 >= 0.01
# el resultado nos indica si la frecuencia del marcador en el total de muestras es igual o mayor a 1%
```

```r
# En este ejemplo seleccionamos solo las columnas (marcadores) con una o mas apariciones
# y creamos el archivo resultante

colnames(matrixTsilv[,1:20])  #lista inicial de marcadores

tot <- frequency(matrixTsilv[,1:20])    # conteo

# tambien analiza la primera columna que no tiene marcadores y por lo tanto la va a eliminar
# pero la volvemos a unir con los sgtes codigos:
final <- matrixTsilv[,1:20]
N <- final[,1]
one <- cbind(N,final[,tot  >= 1 ])    #archivo final editado

dim(one)
one[1:8,]

# se redujo de 19 a 14 marcadores
```
